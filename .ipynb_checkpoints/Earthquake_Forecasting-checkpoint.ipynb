```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.multioutput import MultiOutputRegressor
from sklearn.metrics import mean_squared_error
```

```python
# Load dataset
dataset = pd.read_csv("Earthquake_dataset.csv")
```

```python
print(dataset.head())
```

```python
print(dataset.info())
```

```python
print(dataset.describe())
```

```python
# Data preprocessing
dataset['nst'].fillna(dataset['nst'].mode().iloc[0], inplace=True)
```

```python
dataset['gap'].fillna(dataset['gap'].mode().iloc[0], inplace=True)
```

```python
dataset['rms'].fillna(dataset['rms'].mode().iloc[0], inplace=True)
```

```python
import re
lst = []
for i in dataset['place']:
    new = re.sub('^(.*of )', "", i)
    lst.append(new)
dataset['place'] = lst

df = pd.get_dummies(dataset, columns=['place'])

X = df.drop(['time', 'latitude', 'longitude', 'mag', 'magType', 'dmin', 'net', 'id', 'updated', 'type', 'horizontalError', 'depthError', 'magError', 'magNst', 'status', 'locationSource', 'magSource'], axis=1)
y = df[['mag']]
```

```python
# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=13)
```

```python
# Model training
model = MultiOutputRegressor(RandomForestRegressor(n_estimators=200, max_depth=5, min_samples_leaf=3, random_state=13))
model.fit(X_train, y_train)
```

```python
# Model evaluation
y_pred = model.predict(X_test)
rms = mean_squared_error(y_test, y_pred, squared=False)
print(f"Root Mean Squared Error: {rms}")
```

```python
# Scatter plot for predicted vs true values
plt.figure(figsize=(8,6))
plt.scatter(y_test, y_pred, c='blue')
plt.xlabel("True Magnitude")
plt.ylabel("Predicted Magnitude")
plt.title("True vs Predicted Magnitude")
plt.grid(True)
plt.show()
```

```python
# Residuals Plot
residuals = y_test.values.flatten() - y_pred.flatten()
plt.figure(figsize=(8,6))
sns.histplot(residuals, kde=True)
plt.title("Residuals Distribution")
plt.xlabel("Residuals")
plt.grid(True)
plt.show()
```

```python
# Feature Importance Plot
rf = model.estimators_[0]  # take the first regressor from MultiOutput
importances = rf.feature_importances_
features = X.columns

indices = np.argsort(importances)[::-1]
plt.figure(figsize=(10,8))
sns.barplot(x=importances[indices], y=features[indices])
plt.title("Feature Importances from RandomForest")
plt.show()
```

```python
# Distribution of Target Variable (Magnitude)
plt.figure(figsize=(8,6))
sns.histplot(y, kde=True, bins=30)
plt.title("Magnitude Distribution")
plt.xlabel("Magnitude")
plt.grid(True)
plt.show()
```
